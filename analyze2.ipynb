```python
import numpy
import matplotlib.pyplot as plt
```

```python
amplitude = numpy.pi/4
frequency = 10
phaseOffset = 0

# create motor target angles
targetAngles = numpy.linspace(0, 2*numpy.pi, 1000)
for i, value in enumerate(targetAngles):
    targetAngles[i] = amplitude * numpy.sin(frequency * targetAngles[i] + phaseOffset)

plt.plot(numpy.sin(targetAngles))

plt.xlabel('Angle [rad]')

plt.ylabel('sin(x)')

plt.axis('tight')

plt.show()
```
